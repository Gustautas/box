```python
pos = [14,15]
move_list = list(range(0,15))
re_move_list = move_list.copy()
re_move_list.reverse()
while True:
    #OUT
    for move in move_list:
        pixels.fill((INT,0,0))
        for side in range(0,SN):
            pixels[pos[0]-move+side*SL] = (0,0,INT)
            pixels[pos[1]+move+side*SL] = (0,0,INT)
        pixels.show()
        time.sleep(0.1)
    
    #BOOM
    time.sleep(0.2)
    pixels.fill((0,INT/10,0))
    pixels.show()
    time.sleep(0.5)
    #TODO buildup
    
    #IN
    for move in re_move_list[1:-1]:
        pixels.fill((INT,0,0))
        for side in range(0,SN):
            pixels[pos[0]-move+side*SL] = (0,0,INT)
            pixels[pos[1]+move+side*SL] = (0,0,INT)
        pixels.show()
        time.sleep(0.075)
```
